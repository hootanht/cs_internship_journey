```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
```

# Quick start

```python
X=np.linspace(0,10,100)
Y=np.cos(X)
```

```python
fig,ax=plt.subplots()
ax.plot(X,Y,color='green')

fig.savefig('figure.pdf')
plt.show()
```

# subplots layout

[subplots](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html#matplotlib.pyplot.subplots)

```python
# First create some toy data:
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)
# Create four polar Axes and access them through the returned array
fig, axs = plt.subplots(2, 2, subplot_kw=dict(projection="polar"))
axs[0, 0].plot(x, y)
axs[1, 1].scatter(x, y)
```

```python
gs = mpl.gridspec.GridSpec(8, 39)
ax1 = plt.subplot(gs[:6, :35])
ax2 = plt.subplot(gs[6:, :])

data1 = np.random.rand(6, 35)
data2 = np.random.rand(2, 39)

ax1.imshow(data1)
ax2.imshow(data2)

plt.show()
```

description:
1-initializes a GridSpec object gs with a grid of 8 rows and 39 columns
2-creating subplots

# basic plots

[plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)

```python
x = [1, 2, 3]
y = np.array([[1, 2], [3, 4], [5, 6]])
plt.plot(x, y)
```

[scatter](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html#matplotlib.pyplot.scatter)

```python
# Fixing random state for reproducibility
np.random.seed(19680801)


N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

[bar](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html)

```python
fig,ax=plt.subplots()
ax.bar([1,2,3],[1,2,3])
fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']
ax.bar(fruits, counts, color=bar_colors, label=bar_labels)
ax.legend(loc='upper left')
plt.show()
```
