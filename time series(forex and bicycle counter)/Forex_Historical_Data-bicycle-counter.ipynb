**simple moving average**

In this section we ad moving average columns to our dataframe. but what is moving average? A moving average (MA) is a technical indicator used to identify trends and smooth out price fluctuations. It is calculated by averaging the closing prices of a currency pair over a specified period, such as 5, 10, 20, 50, 100, or 200 days.This averaging process helps to eliminate short-term noise and provide a clearer picture of the underlying trend.

Moving averages are commonly used in Forex trading to:
* Identify trends: An upward trend in a moving average indicates a rise in the price or momentum, while a downward trend indicates a fall.
* Generate buy and sell signals: When the market price crosses the MA line, it can be interpreted as a buy or sell signal.
* Set support and resistance levels: Moving averages can help identify key levels of support and resistance.

```python
data['close'].plot()
```

```python
data['MA1']=data['close'].rolling(window=5).mean()
data['close'].plot()
data['MA1'].plot()
```

```python
data['MA2']=data['close'].rolling(window=10).mean()
data['close'].plot()
data['MA2'].plot()
```

```python
data['MA3']=data['close'].rolling(window=15).mean()
data['close'].plot()
data['MA3'].plot()
```

```python
data['MA4']=data['close'].rolling(window=30).mean()
data['close'].plot()
data['MA4'].plot()
```

```python
data['MA5']=data['close'].rolling(window=100).mean()
data['close'].plot()
data['MA5'].plot()
```

```python
data['MA6']=data['close'].rolling(window=1000).mean()
data['close'].plot()
data['MA6'].plot()
```

**Increasing the window size has several effects:**

* Smoothing: Larger window sizes tend to smooth out the indicator values, reducing the impact of short-term noise and random fluctuations. This can help to identify longer-term trends and reduce false signals.
* Lagging: However, larger window sizes can also lead to a lagging effect, where the indicator values are slower to respond to changes in the market. This can result in delayed entry or exit signals, potentially reducing profitability.

**Visualizing the Data**


 plotting the
raw data

```python
import matplotlib.pyplot as plt

data.plot()
plt.ylabel('Hourly Bicycle Count')
```

The ~150,000 hourly samples are far too dense for us to make much sense of. We can
gain more insight by resampling the data to a coarser grid. Let’s resample by week. All the values within a week will be added together to form a single value for that week.

```python
weekly = data.resample('W').sum()
weekly.plot(style=['-', ':', '--'])
plt.ylabel('Weekly bicycle count')
weekly.head(5)
```

This reveals some trends: as you might expect, people bicycle more in the summer
than in the winter, and even within a particular season the bicycle use varies from
week to week . Further, the effect of the COVID-19 pandemic on commuting patterns is
quite clear, starting in early 2020.


data should be resampled to a daily frequency. The sum function is used to aggregate the values within each day, effectively summing up all the hourly values within a day to get a daily total. Here we’ll examine the 30-day rolling mean
of our data

```python
daily = data.resample('D').sum()
daily.rolling(30, center=True).sum().plot(style=['-', ':', '--'])
plt.ylabel('mean hourly count');
```

**Diggin into the data**

با اینکه تدابیر بالا تونست ما رو به داده های به اصطلاح smoothتری برسونه، اما با اینحا خیلی از روندهای داده از چشم  ما پنهان مونده مثل میانگین ترافیک به عنوان تابعی از زمان روز. که این اطلاعات رو میتونیم با groupby بدست بیاریم:


```python
import numpy as np
by_time = data.groupby(data.index.time).mean()
hourly_ticks = 4 * 60 * 60 * np.arange(6)
by_time.plot(xticks=hourly_ticks, style=['-', ':', '--']);
```

we can see peaks are around 8:00 a.m. and
5:00 p.m.
This is likely evidence of a strong component of commuter traffic crossing
the bridge

We also might be curious about how things change based on the day of the week.
Again, we can do this with a simple groupby

```python
by_weekday = data.groupby(data.index.dayofweek).mean()
by_weekday.index = ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun']
by_weekday.plot(style=['-', ':', '--']);
```
